# Common use of the IAPWSPy package

IAPWSPy is a python package that allows to manipulate thermophysical properties of water, thanks to the IAPWS-95 (International Association for the Properties of Water and Steam : http://www.iapws.org/) experimental correlations, published in [1]. All correlations are currently based on the saturated properties of water and package allow to draw some usual thermodynamical diagrams (Mollier, Clapeyron,...).

[1] Wagner, W. and Pruss, A., *The IAPWS formulation 1995 for the thermodynamic properties of ordinary water substance for general and scientific use*, Journal of Physical and Chemical Reference Data, 31, p. 387, 2002.

```python
from IAPWSPy import *
import numpy as np
import matplotlib.pyplot as pl
```

## Physical constants

All following experimental correlations are built on the two fondamental states of water, corresponding to the **triple point** and the **critical point** :

```python
print('Triple point of water is at a pressure pt = %3.2e bar and a temperature Tt = %3.2f K' % (pt,Tt))
print('Critical point of water is at a pressure pc = %3.2f bar and a temperature Tc = %3.2f K' % (pc,Tc))
```

## Saturation pressure

When water is not mixed with an other component, its physical properties are mainly imposed by the *saturated* state. For a given value of temperature $T$ in $[{\rm K}]$, corresponding saturation pressure is :

```python
# Temperature written in Celcius
T = 80;
# But manipulated in Kelvins
print('Saturation pressure at T = %3.2f°C = %3.2f K is psat = %3.4f bar' % (T,T+273.15,psat(T+273.15)))
```

Same function could be used with a list/array as argument :

```python
# A list of temperature values within the physical range, i.e. between the triple point temperature and the critical one
ListT = np.linspace(Tt,Tc,101)
# And the graph
pl.grid(True)
pl.ylim((pt,pc))
pl.xlim((Tt,Tc))
pl.title(r'Saturation Pressure vs. Temperature',size=18)
pl.xlabel(r'Temperature in $[{\rm K}]$',size=16)
pl.ylabel(r'Pressure in $[{\rm bar}]$',size=16)
pl.semilogy(ListT,psat(ListT),lw=2)
```

Obviously, inverse function exists, that give the saturation temperature for a given pressure :

```python
p = 20.0 # Pressue is in bar here
print('At pressure p = %3.1f bar, saturation temperature is %3.2f K = %3.2f°C' % (p,Tsat(p),Tsat(p)-273.15))
```

And for a list of pressure, within the $T_t < T < T_c$ range 

```python
# List of pressure in bar
Listp = np.linspace(1,100,10)
Tsat(Listp)
```

## Density & specific volume

**Densities** and **specific volumes** of both liquid and vapor phases could be calculated at the saturated state. Argument needed is the temperature :

```python
ListT = np.linspace(Tt,Tc,201)
# Graph of densities evolutions against temperature
pl.grid(True)
pl.title(r'Density vs. Temperature',size=18)
pl.xlabel(r'Temperature in $[{\rm K}]$',size=16)
pl.ylabel(r'Density in $[{\rm kg/m}^3]$',size=16)
pl.semilogy(ListT,liquid_density(ListT),'b-',ListT,vapor_density(ListT),'b--',lw=2.0)
pl.legend((r'Liquid',r'Vapor'),shadow=True,loc=(0.9,0.1))
```

```python
print('At T = %3.2f°C, specific volumes of liquid and vapor phases are respectively vl = %3.3e m3/kg \
and vv = %3.3f m3/kg' % (T,liquid_specific_volume(T+273.15),vapor_specific_volume(T+273.15)))
```

Or starting from given pressures :

```python
vapor_specific_volume(Tsat(Listp))
```

## Enthalpy and internal energy

Specific **enthalpies** and **internal energies** of both phases could be manipulated the very same way :

```python
pl.grid(True)
pl.title(r'Enthalpy vs. Temperature',size=18)
pl.xlabel(r'Temperature in $[{\rm K}]$',size=16)
pl.ylabel(r'Enthalpy in $[{\rm kJ/kg}]$',size=16)
pl.plot(ListT,liquid_enthalpy(ListT),'b-',ListT,vapor_enthalpy(ListT),'b--',lw=2.0)
pl.legend((r'Liquid',r'Vapor'),shadow=True,loc=(0.7,0.1))
```

```python
print('At T = %3.2f°C, specific internal energies of liquid and vapor phases are respectively ul = %3.2f m3/kg \
and uv = %3.2f m3/kg' % (T,liquid_internal_energy(T+273.15),vapor_internal_energy(T+273.15)))
```

## Diagrams drawing

Combining previous correlations could help to draw the usual thermodynamical diagrals of water.

### Mollier's diagram

Mollier's diagram, commonly used in HVAC systems studies, is based on the *specific enthalpy* as x-axis and pressure as y-axis :

```python
# Mollier diagram of water drawn thanks to previous correlations
pl.grid(True)
pl.xlim((0.0,3000))
pl.ylim((pt,2*pc))
pl.title(r'Mollier diagram $(p,h)$',size=18)
pl.xlabel(r'Enthalpy in $[{\rm kJ/kg}]$',size=16)
pl.ylabel(r'Pressure in $[{\rm bar}]$',size=16)
pl.semilogy(liquid_enthalpy(ListT),psat(ListT),'b-',vapor_enthalpy(ListT),psat(ListT),'b--',lw=2)
pl.legend((r'Liquid',r'Vapor'),shadow=True,loc=(0.5,0.1))
# Critical point
pl.scatter([liquid_enthalpy(Tc)],[pc],s=50,marker='o',color='k')
```

### Clapeyron diagram

```python
pl.grid(True)
pl.xlim((1e-3,10.0))
pl.ylim((pt,1.1*pc))
pl.title(r'Clapeyron diagram $(p,v)$',size=18)
pl.xlabel(r'Specific volume $v$ in $[{\rm m}^3/{\rm kg}]$',size=16)
pl.ylabel(r'Pressure $p$ in $[{\rm bar}]$',size=16)
pl.semilogx(liquid_specific_volume(ListT),psat(ListT),'b-',vapor_specific_volume(ListT),psat(ListT),'b--',lw=2)
pl.legend((r'Liquid',r'Vapor'),shadow=True,loc=(0.5,0.5))
# Critical point
pl.scatter([1/rhoc],[pc],s=50,marker='o',color='k')
```
